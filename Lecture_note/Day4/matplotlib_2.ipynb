## 산점도 (Scatter plot)
- `plt.scatter()`
- 옵션
    - `marker` : 모양 설정
    - `c` : 색깔 지정
    - `s` : 점 크기 지정

```python
x = np.random.rand(100)
y = np.random.rand(100)

plt.scatter(x, y)
plt.show()
```

```python
# 점 모양 변경
plt.scatter(x, y, marker = ">")  
plt.show()
```

```python
# 점 색 지정 + 컬러바 추가
colormap = x

plt.scatter(x, y, c = colormap)  
plt.colorbar()
plt.show()
```

```python
# 점 크기 지정
colormap = x
pt_size = (15 * np.random.rand(100))**2

plt.scatter(x, y, c = x*y, s = pt_size)  
plt.colorbar()
plt.show()
```

## 막대 그래프 (bar chart)
- `plt.bar()`
- 옵션
    - `width` : 막대 너비 조절
    - 

```python
x = np.arange(0, 10)
y = np.random.rand(10)

plt.bar(x, y)
plt.show()
```

```python
# 막대 여러개 그리기
z = abs(np.random.randn(10))

plt.bar(x, y, color = "r", width = 0.3, label = "A")
plt.bar(x+0.4, y*z, color = "g", width = 0.3, label = "B")  # x+0.4 : x에 0.4를 더한 위치에 그려라
plt.xlabel("data")
plt.ylabel("value")
plt.legend()
plt.show()
```

```python
# 막대 여러개 그리기
plt.barh(x, y, color = "r", label = "A")
plt.barh(x, y*z, color = "g", label = "B")
plt.xlabel("data")
plt.ylabel("value")
plt.legend()
plt.show()
```

```python
# 막대 여러개 그리기
plt.barh(x, y, color = "r", label = "A")
plt.barh(x, -y*z, color = "g", label = "B")
plt.xlabel("data")
plt.ylabel("value")
plt.legend()
plt.show()
```

# 6. pandas로 그래프 그리기
- `데이터프레임.plot()`을 사용하며 여러 속성을 가지고 있음
- `bar`, `pie`, `hist`, `box`, `scatter` 등
- 기본적으로 열 마다 그래프를 그려줌

## 선 그래프 예시

```python
np.random.seed(0)
df1 = pd.DataFrame(np.random.randn(100, 3),
                   index=pd.date_range('1/1/2020', periods=100),
                   columns=['A', 'B', 'C']).cumsum()
df1.head()
```

```python
df1.plot()
plt.xlabel("Time")
plt.ylabel("Data")
plt.show()
```

## 파이차트 예시
- iris 데이터에서 species의 종류로 파이차트를 그림

```python
import pandas as pd

iris = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
iris.head()
```

```python
df = iris.species.value_counts()
df.plot.pie(autopct='%.2f%%')
plt.legend()
plt.show()
```

## 박스플랏 예시
- iris 데이터에서 연속형 변수 4개에 대한 boxplot

```python
iris.plot.box()
plt.show()
```

```python
iris.boxplot(by = "species")
plt.show()
```

### 막대 그래프 예시
- iris 데이터에서 연속형 변수 4개에 대한 평균을 막대 그래프로 표현

```python
df = iris.groupby("species").mean()
df
```

```python
df.plot.bar(rot = 0)
plt.show()
```
