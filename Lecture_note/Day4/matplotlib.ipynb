## matplotlib에서 한글 사용하기
- 한글을 제목, 범례 등에서 사용하기 위해서는 다음을 실행해주어야함!!

### 1. Windows인 경우
```python
import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
```

### 2. Mac인 경우
```python
import matplotlib.pyplot as plt
plt.rc('font', family = 'AppleGothic')
plt.rcParams['axes.unicode_minus'] = False
```

````python
%matplotlib inline   # 그래프를 노트북 안에 그리기 위해 설정
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-5, 6)
y = x**2

plt.figure()
plt.plot(x, y)
plt.title("한글 테스트")
plt.show()
````

````python
import matplotlib.pyplot as plt

### 한글폰트 설정부분 ###
# Mac인 경우
plt.rc('font', family = 'AppleGothic')
# Windows인 경우
# plt.rc('font', family = 'Malgun Gothic')

plt.rcParams['axes.unicode_minus'] = False
######################

plt.figure()
plt.plot(x, y)
plt.title("한글 테스트")
plt.show()
````
