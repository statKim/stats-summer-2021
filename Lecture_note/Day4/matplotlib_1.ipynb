# 2. matplotlib으로 그래프 그리기
- 그래프를 쌓은 후에, `plt.show()`로 한 번에 보여줌
- 그리는 순서
    1. 데이터 생성
    2. `plt.plot()`으로 그래프 설정 (모양, 색상 등)
    3. 그래프 제목, 축이름, 범례 등 설정 (선택사항)
    4. `plt.show()`로 그래프 그리기

## 그래프 그리기

````python
import numpy as np

# 1. 데이터 생성
x_1 = np.arange(5)
y_1 = x_1

x_2 = np.arange(5)
y_2 = x_2 + 5
````

````python
# 2. 그래프 설정
plt.plot(x_1, y_1)

# 4. 그래프 그리기
plt.show()
````

## 그래프 중첩해서 그리기

````python
# 그래프 2개 설정
plt.plot(x_1, y_1)
plt.plot(x_2, y_2)

# 그래프 그리기
plt.show()
````

# 3. 여러가지 그래프 옵션
- 그래프를 좀 더 예쁘게 그리기 위한 많은 옵션들이 존재 (색, 모양 등)
- 그래프 옵션은 **절대 외우는 것이 아님!!**
- 필요한 옵션은 항상 구글링을 하여 찾는 것임!!

### 색상 설정
- `color` or `c` 옵션
- rgb값, 문자열 색("red", "blue" 등), 축약된 문자열("b"(blue), "g"(green))이 사용가능

````python
plt.plot(x_1, y_1, color = "#100000")
plt.plot(x_2, y_2, c = "r")

plt.show()
````

### 선 모양 설정
- `linestyle` or `ls` 옵션

````python
plt.plot(x_1, y_1, linestyle = "dashed")
plt.plot(x_2, y_2, ls = "dotted")

plt.show()
````

### 마커(marker) 설정
- `marker` 옵션

````python
plt.plot(x_1, y_1, marker = "o")
plt.plot(x_2, y_2, marker = "x")

plt.show()
````

### 제목 추가
- `plt.title(제목)`

````python
plt.plot(x_1, y_1, marker = "o", linestyle = "dashed")
plt.plot(x_2, y_2, marker = "x", linestyle = "dotted")

# 제목 추가
plt.title("선 그래프 2개")

plt.show()
````

### 축 이름 추가
- `plt.xlabel(x축_이름)` : x축 이름 설정
- `plt.ylabel(y축_이름)` : y축 이름 설정

````python
plt.plot(x_1, y_1, marker = "o", linestyle = "dashed")
plt.plot(x_2, y_2, marker = "x", linestyle = "dotted")

plt.title("선 그래프 2개")

# 축 이름 추가
plt.xlabel("x축")
plt.ylabel("y축")

plt.show()
````

### 범례 추가
- `plt.legend(loc)`
- `loc` 옵션은 범례의 위치를 지정
    - ex) "upper right", "lower left", "right", "center" 등
- 또한 각 선마다의 label값을 `plt.plot()`에 옵션으로 추가해주어야 함!!

````python
plt.plot(x_1, y_1, marker = "o", linestyle = "dashed", label = "파랑")
plt.plot(x_2, y_2, marker = "x", linestyle = "dotted", label = "주황")

plt.title("선 그래프 2개")
plt.xlabel("x축")
plt.ylabel("y축")

# 범례 추가
plt.legend(loc = "lower right")

plt.show()
````

### 축 범위 설정
- `plt.xlim(x축_시작값, x축_끝값)`
- `plt.ylim(y축_시작값, y축_끝값))`

````python
plt.plot(x_1, y_1, marker = "o", linestyle = "dashed", label = "파랑")
plt.plot(x_2, y_2, marker = "x", linestyle = "dotted", label = "주황")

plt.title("선 그래프 2개")
plt.xlabel("x축")
plt.ylabel("y축")
plt.legend(loc = "lower right")

# 축 범위 설정
plt.xlim(-3, 6)
plt.ylim(-2, 11)

plt.show()
````

### 그리드 추가
- `plt.grid(True)`

````python
plt.plot(x_1, y_1, marker = "o", linestyle = "dashed", label = "파랑")
plt.plot(x_2, y_2, marker = "x", linestyle = "dotted", label = "주황")

plt.title("선 그래프 2개")
plt.xlabel("x축")
plt.ylabel("y축")
plt.legend(loc = "lower right")
plt.xlim(-3, 6)
plt.ylim(-2, 11)

# 그리드 추가
plt.grid(True)

plt.show()
````

### 축의 표시 값 설정
- `plt.xticks(x축_표시값)`
- `plt.yticks(y축_표시값)`

````python
plt.plot(x_1, y_1, marker = "o", linestyle = "dashed", label = "파랑")
plt.plot(x_2, y_2, marker = "x", linestyle = "dotted", label = "주황")

plt.title("선 그래프 2개")
plt.xlabel("x축")
plt.ylabel("y축")
plt.legend(loc = "lower right")
plt.xlim(-3, 6)
plt.ylim(-2, 11)
plt.grid(True)

# 축 표시 값 설정
plt.xticks([-3, 0, 3, 6])
plt.yticks([-2, 4, 10])

plt.show()
````

# 4. 서브 플랏 그리기
- 한 화면에 그래프를 여러개 그리기 위해 사용
- 여러 그래프를 각 지정한 위치마다 그릴 수 있음
- 사용방법
```python
plt.subplot(행_개수, 열_개수, 위치)
<그래프 코드>
```
- 위치의 경우, 이전의 인덱싱 방법과 다르게 0이 아닌 1부터 시작

## 2 x 1 형태의 subplot
- `plt.subplot(2, 1, 위치)`

````python
ax1 = plt.subplot(2, 1, 1)
plt.plot(x_1, y_1, marker = "o", color = "red", linestyle = "dashed")

ax2 = plt.subplot(2, 1, 2)
plt.plot(x_2, y_2, marker = "x", color = "blue", linestyle = "dotted")

plt.tight_layout()
plt.show()
````

## 2 x 2 형태의 subplot
- `plt.subplot(2, 2, 위치)` or `plt.subplot(22위치)`(comma 없이 해도 가능)
- 가로 방향부터 순서대로 그려짐

````python
np.random.seed(0)

plt.subplot(2, 2, 1)
plt.plot(np.random.rand(5))
plt.title("axes 1")

plt.subplot(2, 2, 2)
plt.plot(np.random.rand(5))
plt.title("axes 2")

plt.subplot(2, 2, 3)
plt.plot(np.random.rand(5))
plt.title("axes 3")

plt.subplot(2, 2, 4)
plt.plot(np.random.rand(5))
plt.title("axes 4")

plt.tight_layout()
plt.show()
````

#### 다음과 같이 행과 열의 인덱스로 지정하여서도 그릴 수 있음

````python
fig, axes = plt.subplots(2, 2)

np.random.seed(0)
axes[0, 0].plot(np.random.rand(5))
axes[0, 0].set_title("axes 1")

axes[0, 1].plot(np.random.rand(5))
axes[0, 1].set_title("axes 2")

axes[1, 0].plot(np.random.rand(5))
axes[1, 0].set_title("axes 3")

axes[1, 1].plot(np.random.rand(5))
axes[1, 1].set_title("axes 4")

plt.tight_layout()
plt.show()
````
